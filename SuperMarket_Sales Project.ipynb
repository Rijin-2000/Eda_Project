```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r'C:\Users\Admin\Desktop\SuperMarket Sales.csv')
```

```python
for i in df:
    j=i.replace(' ','_')
    
    df=df.rename(columns={i:j.capitalize()})
```

```python
missing_values=df.isnull().sum()
print("missing values:")
print(missing_values)
```

### 1) Top Selling Product Lines

```python
# What are the top selling product lines?

df['Product_line'].value_counts().nlargest()
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_line',y='Quantity',palette='Set2',data=df)
plt.title('Top selling product lines')
plt.xlabel('Product Line')
plt.ylabel('Total Quantity')
plt.show()
```

#### The analysis identifies the most popular product lines by sales quantity and visually represents this data in a bar chart. This helps in understanding which product lines are the most successful, aiding in inventory and marketing strategies.

### 2) Top Rating Product Lines

```python
# What are top rating product lines?

df['Rating'].value_counts().nlargest(1)
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_line',y='Rating',palette='Set2',data=df)
plt.title('Top rating product lines')
plt.xlabel('Product Line')
plt.ylabel('Rating')
plt.show()
```

#### The analysis identifies the product lines that have received the highest average ratings from customers and visually represents this data in a bar chart. This helps in understanding which product lines are the most appreciated by customers, aiding in quality control and product development strategies.

### 3) Least Selling Product Lines

```python
# What are the least selling product lines?

sales_by_product_line = df.groupby('Product_line')['Total'].sum().reset_index()
least_selling = sales_by_product_line.sort_values(by='Total').head(5)
plt.figure(figsize=(12, 6))
sns.barplot(data=least_selling, x='Product_line', y='Total', palette='Set2')
plt.title('Least Selling Product Lines')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

#### The analysis identifies and visualizes the product lines with the lowest sales, providing a clear graphical representation of the least performing segments in terms of sales.

### 4) Highest prefered payment mode

```python
# Which payment mode has the highest preference?
 
payment_mode_counts = df['Payment'].value_counts().reset_index()
payment_mode_counts.columns = ['Payment', 'count']
print(payment_mode_counts)
plt.figure(figsize=(8,8))
plt.pie(payment_mode_counts['count'], labels=payment_mode_counts['Payment'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2', len(payment_mode_counts)))
plt.title('Preferred Payment Modes')
plt.show()
```

#### The analysis identifies and visualizes the preferred payment modes among customers, with the pie chart clearly showing the relative popularity of each mode. This helps in understanding customer payment preferences, which can inform business decisions related to payment processing and customer convenience.

### 5) Top 5 Highest Sales Dates

```python
# Which is the top 5 highest sales date?

sales_by_date = df.groupby('Date')['Unit_price'].sum().reset_index()
top_5_sales_dates = sales_by_date.sort_values(by='Unit_price', ascending=False).head(5)
print(top_5_sales_dates)
plt.figure(figsize=(8, 8))
plt.pie(top_5_sales_dates['Unit_price'], labels=top_5_sales_dates['Date'], autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Top 5 Sales Dates')
plt.show()
```

#### The top five highest sales dates are identified and visualized in a pie chart, indicating the days with the most significant revenue contributions. This analysis helps in understanding peak sales periods, which can be critical for optimizing inventory and marketing strategies.

### 6) Least 5 Sales Dates

```python
# Which is the least 5 sales date?

sales_by_date = df.groupby('Date')['Unit_price'].sum().reset_index()
least_5_sales_dates = sales_by_date.sort_values(by='Unit_price', ascending=True).head(5)
print(least_5_sales_dates)
plt.figure(figsize=(8, 8))
plt.pie(least_5_sales_dates['Unit_price'], labels=least_5_sales_dates['Date'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(least_5_sales_dates)))
plt.title('Lowest Sales Dates')
plt.show()
```

#### The analysis reveals the five dates with the lowest sales, visualized in a pie chart, highlighting periods of minimal revenue generation. Understanding these low-sales dates can aid in identifying potential issues or opportunities for improvement in sales strategies during these times.

### 7) Top Customer Type for Product Line

```python
# Which type of customers is buying more product line?

agg_data = df.groupby(['Customer_type', 'Product_line']).sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=agg_data, x='Customer_type', y='Quantity', hue='Product_line', palette='Set2')
plt.title('Quantity of Product Lines Purchased by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Quantity Purchased')
plt.legend(title='Product Line')
plt.show()
```

#### The bar plot shows the quantity of different product lines purchased by each customer type, indicating which customer group contributes more to specific product categories. This insight helps in tailoring marketing efforts and inventory management to target the most profitable customer segments effectively.

### 8) City with the Most Customers

```python
# Most of the customers were in which city?

city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Number_of_Customers']
print(city_counts)
plt.figure(figsize=(8, 8))
plt.pie(city_counts['Number_of_Customers'], labels=city_counts['City'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2', len(city_counts)))
plt.title('Number of Customers per City')
plt.show()
```

#### The analysis indicates the distribution of customers across various cities, with the highest concentration of customers being in the top city identified by the data. This information is crucial for understanding market presence and planning location-specific marketing strategies to maximize customer reach.

### 9) Top 10 Highest Tax Rates Based on Price

```python
# Which is the 10 highest tax rate based on price

sorted_df = df.sort_values(by='Tax_5%', ascending=False)
top_10_tax_rates = sorted_df.head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_tax_rates, x='Tax_5%', y='Unit_price', palette='Set2')
plt.title('Top 10 Highest Tax Rates vs Unit Price')
plt.xlabel('Tax Rate (%)')
plt.ylabel('Unit Price ($)')
plt.legend()
plt.show()
```

#### The bar plot showcases the top 10 highest tax rates based on unit price, highlighting the correlation between high-priced items and their respective tax contributions. This insight can be used to analyze the impact of taxation on high-value products and inform pricing strategies accordingly.

### 10) Least Preferred Payment Mode

```python
# Which payment mode has the least preference ?

payment_counts = df['Payment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Payment Mode Distribution')
plt.show()
least_preferred_mode = payment_counts.idxmin()
print('The payment mode with the least preference is: {least_preferred_mode}')
```

#### The pie chart illustrates the distribution of different payment modes, identifying the mode with the least preference among customers. Understanding this least preferred payment method helps businesses in assessing customer preferences and potentially reconsidering the offered payment options to enhance customer satisfaction.

### 11) Highest saled branch

```python
# Which branch has the highest sales ?

sales_by_branch = df.groupby('Branch').sum().reset_index()
highest_sales_branch = sales_by_branch.loc[sales_by_branch['Unit_price'].idxmax()]
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_by_branch, x='Branch', y='Unit_price', palette='viridis')
plt.title(f"Branch with Highest Sales: {highest_sales_branch['Branch']} (Total Sales: {highest_sales_branch['Unit_price']})")
plt.xlabel("Branch")
plt.ylabel("Total Unit Price")
plt.show()
```

#### This analysis illustrates the total sales across branches, showcasing the highest performing branch. Through visual representation, it facilitates quick insights into branch-wise sales distribution and identifies the top revenue-generating branch.

### 12) number of unique customers

```python
# How many unique customers are there ?

unique_customers = df['Invoice_id'].nunique()
plt.figure(figsize=(8, 6))
sns.barplot(x=['Unique Customers', 'Others'], y=[unique_customers, len(df) - unique_customers], palette="Set2")
plt.title("Distribution of Unique Customers")
plt.ylabel("Count")
plt.show()
```

#### This analysis reveals the distribution of unique customers within the dataset, emphasizing the prevalence of repeat transactions over individual invoices. By visualizing this relationship, it underscores the significance of customer retention and engagement strategies in sustaining business growth.

### 13) Distribution of customer rating

```python
# What is the distribution of customer rating ?

rating_counts = df['Rating'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=rating_counts, y=rating_counts, palette="Set2")
plt.title("Distribution of Customer Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

#### This barplot depicts the distribution of customer ratings across different categories. It provides insight into the frequency of ratings, highlighting potential areas of customer satisfaction or areas requiring improvement based on the concentration of ratings.

### 14)  Average gross income per sale

```python
# What is the average gross income per sale ?

average_income_per_sale = df['Gross_income'].mean()
plt.figure(figsize=(8, 8))
plt.pie([average_income_per_sale], labels=['Average Gross Income per Sale'], colors=['skyblue'], autopct='%1.1f%%')
plt.title("Average Gross Income per Sale")
plt.show()
```

#### The pie chart illustrates the proportion of the average gross income per sale within the dataset, highlighting its relative share of the total. With an average income of X%, it provides a snapshot of the distribution of earnings across sales, aiding in understanding the central tendency of gross income in the dataset.

### 15) Sales compare between members and normal customers

```python
# How do sales compare between members and normal customers ?

total_sales_members = df[df['Customer_type'] == 'Member']['Gross_margin_percentage'].sum()
total_sales_normal = df[df['Customer_type'] == 'Normal']['Gross_margin_percentage'].sum()
plt.figure(figsize=(8, 8))
sns.set_style("whitegrid")
sns.set_palette("pastel")
plt.title("Comparison of Sales between Members and Normal Customers")
plt.pie([total_sales_members, total_sales_normal], labels=['Members', 'Normal Customers'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

#### This pie chart compares the total gross margin percentages between members and normal customers, highlighting potential differences in profitability. Members appear to contribute a higher proportion of the total gross margin compared to normal customers, suggesting the significance of their contributions to overall sales revenue.

### 16) Sales trend over time

```python
# What is the sales trend over time ?

df['Date'] = pd.to_datetime(df['Date'])
sales_trend = df.groupby(df['Date'].dt.to_period('M'))['Total'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_trend,palette='Set2')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

#### The sales trend over time shows a clear pattern of monthly sales, allowing us to identify periods of high and low sales activity. This information is crucial for strategic planning and optimizing inventory management.

### 17) Highest transacted city

```python
# Which city has the highest number of sales transactions ?

city_sales = df.groupby('City')['Total'].count().reset_index()
city_sales.rename(columns={'Total': 'Number of Transactions'}, inplace=True)
plt.figure(figsize=(8, 8))
plt.pie(city_sales['Number of Transactions'], labels=city_sales['City'], autopct='%1.1f%%', startangle=140)
plt.title('Number of Sales Transactions by City')
plt.axis('equal')
plt.show()
```

#### The pie chart illustrates the distribution of sales transactions across various cities, highlighting the city with the highest number of transactions. This visualization helps in quickly identifying which cities contribute most significantly to the overall sales volume.
